```python
# Librer√≠as necesarias
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
import numpy as np

X = tabla[['pib_per_capita_dolares_corrientes', 
           'tcn_pesos_dolares', 
           'expectativa_inflacion_promedio', 
           'Riesgo pa√≠s', 
           'Reservas_BCRA_millones']]

y = tabla['pib_precios_corrientes']

modelo = LinearRegression()
kfold = KFold(n_splits=5, shuffle=True, random_state=42)


r2_scores = cross_val_score(modelo, X, y, cv=kfold, scoring='r2')
print("R¬≤ por fold:", np.round(r2_scores, 3))
print("Promedio R¬≤:", round(r2_scores.mean(), 3))

mse_scores = -cross_val_score(modelo, X, y, cv=kfold, scoring='neg_mean_squared_error')
print("ECM por fold:", np.round(mse_scores, 2))
print("Promedio ECM:", round(mse_scores.mean(), 2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

r2_scores = [0.909, 0.995, 0.998, 0.992, 0.982]

plt.figure(figsize=(8, 5))
folds = [f'Fold {i+1}' for i in range(len(r2_scores))]

plt.bar(folds, r2_scores, color='mediumseagreen')

mean_r2 = np.mean(r2_scores)
plt.axhline(y=mean_r2, color='red', linestyle='--', label=f'Promedio R¬≤ = {round(mean_r2, 4)}')


plt.ylim(0.90, 1.00)
plt.title('R¬≤ por Fold - Validaci√≥n Cruzada')
plt.ylabel('R¬≤')
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.show()
```
