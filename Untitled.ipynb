```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Variables explicativas y dependiente
X = tabla[['pib_per_capita_dolares_corrientes',
           'tcn_pesos_dolares',
           'expectativa_inflacion_promedio',
           'Riesgo país',
           'Reservas_BCRA_millones']]

y = tabla['icg_nivel_general']

# Configurar K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Listas para almacenar resultados
r2_afuera = []
r2_adentro = []

# Validación cruzada
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    modelo = LinearRegression()
    modelo.fit(X_train, y_train)

    y_pred_train = modelo.predict(X_train)
    y_pred_test = modelo.predict(X_test)

    r2_adentro.append(r2_score(y_train, y_pred_train))
    r2_afuera.append(r2_score(y_test, y_pred_test))

# Resultados promedio
print("R² afuera (promedio):", np.mean(r2_afuera))
print("R² adentro (promedio):", np.mean(r2_adentro))

```

```python
# Librerías necesarias
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
import numpy as np

# Selección de variables explicativas y dependiente
X = tabla[['tcn_pesos_dolares', 'Riesgo país']]
y = tabla['icg_nivel_general']

# Configurar K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Listas para almacenar métricas
r2_train_scores = []
r2_test_scores = []
ecm_train_scores = []
ecm_test_scores = []

# Validación cruzada
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    modelo = LinearRegression()
    modelo.fit(X_train, y_train)

    y_pred_train = modelo.predict(X_train)
    y_pred_test = modelo.predict(X_test)

    r2_train_scores.append(r2_score(y_train, y_pred_train))
    r2_test_scores.append(r2_score(y_test, y_pred_test))

    ecm_train_scores.append(mean_squared_error(y_train, y_pred_train))
    ecm_test_scores.append(mean_squared_error(y_test, y_pred_test))

# Promedios
print("Coeficientes del último modelo:", modelo.coef_)
print("Intercepto del último modelo:", modelo.intercept_)
print("R² promedio entrenamiento (adentro):", round(np.mean(r2_train_scores), 4))
print("R² promedio prueba (afuera):", round(np.mean(r2_test_scores), 4))
print("ECM promedio entrenamiento:", round(np.mean(ecm_train_scores), 2))
print("ECM promedio prueba:", round(np.mean(ecm_test_scores), 2))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Definición de X (variables explicativas) e y (variable objetivo)
X = tabla[['tcn_pesos_dolares', 'Riesgo país']]
y = tabla['icg_nivel_general']

# Configurar K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

r2_scores = []
ecm_scores = []

# Validación cruzada
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    modelo = LinearRegression()
    modelo.fit(X_train, y_train)

    y_pred = modelo.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    ecm_scores.append(mean_squared_error(y_test, y_pred))

# Promedios de desempeño
print("R² promedio:", np.mean(r2_scores))
print("ECM promedio:", np.mean(ecm_scores))

# Gráfico de la última iteración (opcional)
plt.figure(figsize=(7, 7))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', label='Ideal (y = x)')
plt.xlabel('Valor real de ICG')
plt.ylabel('Valor predicho por el modelo')
plt.title('Regresión Lineal Múltiple: Valores reales vs predichos (último fold)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
